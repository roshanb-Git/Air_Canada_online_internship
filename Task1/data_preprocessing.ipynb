```python
#Data Preprocessing
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

# Initialize the Porter Stemmer
ps = PorterStemmer()

# List to hold the cleaned reviews
corpus = []

for i in range(len(df)):
    # Convert to lowercase and remove "✅Trip Verified|" or "Not Verified|"
    review = df['Review'][i].lower()
    review = re.sub(r'^(✅trip verified\| |not verified\| )', '', review)
    
    # Remove non-alphabetic characters
    review = re.sub('[^a-zA-Z]', ' ', review)
    
    # Split the review into words
    review = review.split()
    
    # Stemming and removing stopwords
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    
    # Join the cleaned words back into a single string
    review = ' '.join(review)
    
    # Add the cleaned review to the corpus
    corpus.append(review)
```

```python
#store the cleaned data

df['Cleaned_review'] = corpus

df.head()
```

```python
#Save the DatFrame with cleaned review

df.to_csv('cleaned_air_canada_reviews.csv' , index=False)
```
